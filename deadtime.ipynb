```python
%matplotlib inline
import riesling as rl
!rm dt*.h5
```

```python
settings='--matrix=64 --snr=100 --channels=4 --shepp_logan'
!riesling phantom dt-0 {settings}
!riesling phantom dt-1 {settings} --gap=4 --lores=8
!riesling phantom dt-2 {settings} --gap=4
```

```python
display(rl.plots.traj2d('dt-0.h5', sl_read=-1))
display(rl.plots.traj2d('dt-1.h5', sl_read=-1))
display(rl.plots.traj2d('dt-2.h5', sl_read=-1))
```

```python
!riesling traj --sdc=pipenn dt-0.h5
!riesling traj --sdc=pipenn dt-1.h5
!riesling traj --sdc=pipenn dt-2.h5
display(rl.images.single('dt-0-traj.h5', title='Gap = 0', dset='traj-image'))
display(rl.images.single('dt-1-traj.h5', title='With WASPI', dset='traj-image'))
display(rl.images.single('dt-2-traj.h5', title='Without WASPI', dset='traj-image'))
```

```python
!riesling recon --rss --sdc=pipenn dt-0.h5
!riesling recon --rss --sdc=pipenn dt-1.h5
!riesling recon --rss --sdc=pipenn dt-2.h5
display(rl.images.single('dt-0-recon.h5', title='Gap = 0'))
display(rl.images.single('dt-1-recon.h5', title='With WASPI'))
display(rl.images.single('dt-2-recon.h5', title='Without WASPI'))
```
